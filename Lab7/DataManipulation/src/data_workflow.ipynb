<h5 style="font-style:italic; font-size:16px;">Exercise 3 – Split concatenated CSV into weekly files</h5>

```python
grouped_data['Date'] = pd.to_datetime(grouped_data['Date'])
grouped_data = grouped_data.sort_values('Date')
```

```python
start_date = grouped_data['Date'].min()
end_date = grouped_data['Date'].max()
```

```python
current_start = start_date

while current_start <= end_date:
    current_end = current_start + pd.Timedelta(days=6)
    week_data = grouped_data[(grouped_data['Date'] >= current_start) & (grouped_data['Date'] <= current_end)]
    
    if not week_data.empty:
        file_name = f"{current_start.strftime('%Y%m%d')}_{current_end.strftime('%Y%m%d')}.csv"
        week_data.to_csv(f"C:\\Users\\Costea\\VisualStudioProjects\\DeclarativeProgramming\\Lab7\\DataManipulation\\data\\weekly_files\\{file_name}", index=False)
    
    current_start = current_end + pd.Timedelta(days=1)
```

<h5 style="font-style:italic; font-size:16px;">Exercise 4 – Retrieve rows for a given date</h5>

```python
grouped_data['Date'] = pd.to_datetime(grouped_data['Date'], format='%Y-%m-%d')
```

```python
def get_rows_for_date(target_date):
    try:
        target_date = pd.to_datetime(target_date)
    except Exception:
        print("Error: target_date must be a valid date")
        return None
    result = grouped_data[grouped_data['Date'] == target_date]
    if result.empty:
        print(f"No data available for date {target_date.strftime('%Y-%m-%d')}")
        return None
    return result
```

```python
date_to_check = "2025-12-02"
rows = get_rows_for_date(date_to_check)
print(rows)
```

<h5 style="font-style:italic; font-size:16px;">Exercise 5 – Retrieve rows for a given date range</h5>

```python
grouped_data['Date'] = pd.to_datetime(grouped_data['Date'])
```

```python
def get_rows_for_week(start_date: datetime):
    start = pd.to_datetime(start_date)
    end = start + timedelta(days=6) 
    mask = (grouped_data['Date'] >= start) & (grouped_data['Date'] <= end)
    week_rows = grouped_data.loc[mask]
    if not week_rows.empty:
        return week_rows
    return None
```

```python
start_week = datetime(2025, 12, 2)
week_data = get_rows_for_week(start_week)
if week_data is not None:
    print(f"Data for week starting {start_week.strftime('%Y-%m-%d')}:")
    print(week_data)
else:
    print("No data available for this week")
```
