<h5 style="font-style:italic; font-size:16px;">Exercise 2 - Convert seconds to hh:mm:ss</h5>

```python
total_seconds = random.randint(0, 10000)
print("Total seconds:", total_seconds)
```

<h5 style="font-style:italic; font-size:16px;">Exercise 3 - Float operations with lambda and map</h5>

```python
operations = [
    lambda x, y: x + y,
    lambda x, y: x - y
]
```

```python
a = 12.5
b = 7.3

for func in operations:
    print(func(a, b))
```

```python
values = [1, 2, 3]
nums = [4, 5, 6]

result = list(map(lambda x, y: x + y, values, nums))
print(result)
```

<h5 style="font-style:italic; font-size:16px;">Exercise 4 - Display distinct integers using filter</h5>

```python
values = [1, 2, 3, 2, 4, 1, 5, 3]

seen = set()

distinct_values = list(filter(lambda x: x not in seen and not seen.add(x), values))
print(distinct_values)
```

```python
values = [1, 2, 3, 2, 4, 1, 5]

unique_values = set(values)
print(unique_values)
```

<h5 style="font-style:italic; font-size:16px;">Exercise 5 - Count character occurrences with reduce</h5>

```python
def count_char(string, char):
    return reduce(lambda acc, c: acc + (c == char), string, 0)
```

```python
text = "hello world"
character = "l"

occurrences = count_char(text, character)
print(f"Character '{character}' occurs {occurrences} times in '{text}'")
```

```python
def count_char(string, char):
    count = 0
    for c in string:
        if c == char:
            count += 1
    return count
```

```python
text = "hello world"
character = "l"

occurrences = count_char(text, character)
print(f"Character '{character}' occurs {occurrences} times in '{text}'")
```

<h5 style="font-style:italic; font-size:16px;">Exercise 6 - Concatenate any number of strings</h5>

```python
def concat_strings(*args):
    result = ""
    for s in args:
        result += s
    return result
```

```python
res = concat_strings("Apple", " ", "Rocket", " ", "Sunny", " ", "Laptop", " ", "Blue", " ", "Elephant")
print(res)
```

<h5 style="font-style:italic; font-size:16px;">Exercise 7 - Lambda functions on strings</h5>

```python
funcs = [
    lambda s: s[1] if len(s) > 1 else None,          
    lambda s: s.upper(),                              
    lambda s: s.lower(),                             
    lambda s: s[0] if s else None,                  
    lambda s: s[-1] if s else None,                  
    lambda s: len(s),                              
    lambda s, c: s.count(c),                         
    lambda s, c: s.find(c)                           
]
```

```python
sample_string = "Python"
char_to_find = "t"

print("Second character:", funcs[0](sample_string))
print("Uppercase:", funcs[1](sample_string))
print(f"Position of '{char_to_find}':", funcs[2](sample_string, char_to_find))
```

<h5 style="font-style:italic; font-size:16px;">Exercise 8 - Calculate average of three grades with default values</h5>

```python
def average_grade(grade1=4, grade2=4, grade3=4):
    return (grade1 + grade2 + grade3) / 3
```

```python
print("Average:", average_grade())
print("Average:", average_grade(6, 8))
print("Average:", average_grade(7, 9, 10))
print("Average:", average_grade(grade2=10, grade1=8))
```

<h5 style="font-style:italic; font-size:16px;">Exercise 9 - Recursive sum of first n natural numbers</h5>

```python
def sum_natural(n):
    if n == 0:
        return 0
    return n + sum_natural(n - 1)
```

```python
n = 10
result = sum_natural(n)
print(f"Sum of first {n} natural numbers is: {result}")
```

```python
def sum_natural_iterative(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
```

```python
n = 10
result = sum_natural_iterative(n)
print(f"Sum of first {n} natural numbers is: {result}")
```
